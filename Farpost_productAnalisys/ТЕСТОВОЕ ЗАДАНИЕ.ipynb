## 1. Построение дизайна АБ-теста
---
Дизайн АБ-теста, позволяющий оценить эффект от изменения верстки объявления на сайте Фарпост (раздел Автозапчасти). 

**Основная фича:** Вместо списка делаем плиточный вид. 

![objects.png](objects.png)

Наш АБ-тест будет рассматривать два варианта отображения товаров раздела "Автозапчасти": 
- с помощью списка, как это реализованно сейчас(вариант А);
- с помощью плитки, что я отобразила в своем представлении на картинке выше(вариант Б).

Обозначим **основную цель**, для которой мы запускаем наш АБ-тест: увеличить количество покупок клиентов, количество просмотров и активных пользователей в разделе "Автозапчасти", благодаря изменению вида каталога товаров. Другими словами, вид каталога товаров должен привлекать покупателей.

Вижу цель — не вижу препятствий!

### 2. Определим метрики
---

- #### Основные метрики:

**Конверсия в покупку**: Это процент людей, которые совершили целевое действие. Например, купили товар, перейдя по объявлению. 

**CTR (Click Through Rate)**: Показывает соотношение числа кликов на объявление к числу показов. 

**Average session length**: Средняя длительность одного сеанса. Сколько времени пользователь тратит на взаимодействие с приложением или сайтом.

- #### Дополнительные метрики:

**Глубина просмотра (Pages per Session)**: Если новый дизайн вызывает у пользователей большее желание просматривать другие объявления, эта метрика покажет, насколько улучшилась вовлеченность.

**Количество добавлений в корзину**: Показывает соотношение количества добавлений объявления в корзину к числу показов. 

**Количество добавлений в избранное**: Показывает соотношение количества добавлений объявления в корзину к числу показов. 

- #### Защитные метрики:

**Bounce rate (Показатель отказов)**: Это метрика, которая покажет, сколько пользователей ушло с сайта после просмотра только одной страницы.

### 3. Формируем гипотезу
---

**Нулевая гипотеза**: Изменения от фичи не принесут ожидаемых результатов. Если мы изменим верстку каталога товаров на плиточный вид, это никак не скажется на увеличении основных и дополнительных метрик.

**Альтернативная гипотеза**: Если мы изменим верстку каталога товаров на плиточный вид, мы сможем добиться увеличения основных и дополнительных метрик и увеличим интерес покупателей к товарам.

### 4. Определяем сегмент АВ теста и критерии для сплитирования пользователей.
---

Чтобы эксперимент прошел объективно, выборка должна быть репрезентативной.

Нужно определить две таких группы — экспериментальную и контрольную. Одной аудитории мы показываем вариант сайта A, другой — вариант B.

Таким образом, пользователи должны быть случайным образом распределены на равные две группы:

- **Тестовая группа:** Показываются объявления в плиточном виде.

У нас есть целевая аудитория для выборки, а какой размер выборки нам нужен?

Данный вопрос является важным для обсуждения. Мы можем определить размер самостоятельно, используя специальную формулу, или же мы можем воспользоваться готовым калькулятором: https://www.optimizely.com/sample-size-calculator/#/?conversion=30&effect=5&significance=95

- **Минимальное изменение в конверсии, которое мы хотим обнаружить:** наименьшее изменение в проценте конверсии в покупку, которое для нас имеет практическую значимость и которое мы хотите иметь возможность зафиксировать с помощью теста. Примем это значение за 5%.

- **Уровень значимости:** вероятность отвергнуть нулевую гипотезу, когда она верна. Обычно используется значение 0.05, что соответствует 5%.


Итак, зная эти параметры, попробуем определить размер выборки с помощью готового калькулятора.
Получим размер выборки - 14000 per variation (что означает 14000 на одну группу тестирования)

### 5. Определение длительности теста

Как мы можем отметить, данные графики не имеют нормального распределения. Мы можем попробовать привести их к нормальному виду с помощью следующих методов:

- Логарифмическое преобразование
- Масштабирование StandardScaler
- Масштабирование MinMaxScaler
- Взятие квадратного корня

Для примера попробуем преобразовать одну из метрик - session_length. Нормальность распределения будем проверять используя тест на нормальность (Шапиро-Уилка)

```python
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler

metric = 'session_length'

log_transformed = np.log1p(df[metric])

scaler = StandardScaler()
standard_scaled = scaler.fit_transform(df[[metric]])

minmax_scaler = MinMaxScaler()
minmax_scaled = minmax_scaler.fit_transform(df[[metric]])

sqrt_transformed = np.sqrt(df[metric])

def plot_histograms_and_normality(data, title=""):
    plt.figure(figsize=(15, 5))
    for i, column in enumerate(data.columns, 1):
        plt.subplot(1, len(data.columns), i)
        sns.histplot(data[column], bins=30, kde=True)
        plt.title(f'{column} - {title}')
        
        # Шапиро-Уилка
        stat, p_value = stats.shapiro(data[column])
        print(f'{column} - {title} - p-value: {p_value}')
        if p_value < 0.05:
            print(f'{column} не распределено нормально\n')
        else:
            print(f'{column} имеет нормальное распределение\n')

    plt.tight_layout()
    plt.show()

df_transformed = pd.DataFrame({
    'Логарифмическое преобразование': log_transformed,
    'Standard Scaler': standard_scaled.flatten(),
    'MinMax Scaler': minmax_scaled.flatten(),
    'Взятие квадратного корня': sqrt_transformed
})

plot_histograms_and_normality(df_transformed)
```

Как мы можем заметить, данные преобразования не помогли нам привести данные к нормальному виду, а потому мы не сможем использовать t-критерий Стьюдента для проверки гипотез. Однако, существует еще один тест, который может нам помочь.

Вместо t-критерия Стьюдента, который требует нормальности данных, мы применим тест Манна-Уитни (U-test). Это непараметрический тест, который сравнивает медианы двух независимых групп и не требует нормальности распределений.

##### 6.2 Выбор и реализация способа рассчета 

Рассмотрим оценку тестовой и контрольной группы с помощью оценок метрик, используя тест Манна-Уитни (U-test). 

Зная p-value мы можем сделать следующие выводы:

- Если p-value меньше 0.05, то вероятность того, что различия между группами случайны, очень мала. В этом случае мы отвергаем нулевую гипотезу (H0) и принимаем альтернативную гипотезу (H1), что различия действительно существуют.
- Если p-value больше или равно 0.05, то различия статистически незначимы, и мы не отвергаем нулевую гипотезу. Это означает, что нет оснований для утверждения о различиях между группами.

```python
from scipy import stats

metrics = ['CTR', 'conversion', 'session_length', 'added_to_cart']

for metric in metrics:
    control_group = df_test[df_test['group'] == 'control'][metric]
    test_group = df_test[df_test['group'] == 'test'][metric]
    
    u_stat, p_value = stats.mannwhitneyu(control_group, test_group)
    
    print(f"{metric}: U-статистика = {u_stat:.4f}, p-значение = {p_value:.4f}")
    
    if p_value < 0.05:
        print(f"Различия по метрике {metric} статистически значимы (p < 0.05).")
    else:
        print(f"Различия по метрике {metric} незначимы (p >= 0.05).")
```

В конце своего исследования хочу отметить, что при проведении любого АБ-теста стоит учитывать и иные факторы, которые сложно продемонстрировать в рамках лишь синтетических данных, но которые могут повлиять на результаты тестирования.

##### 7.1 Описание дальнейшей работы с фичей проверяемой в рамках теста, если одна из основных метрик значимо ухудшилась, а другая значимо улучшилась. 


В некоторой ситуации АБ-теста может произойти случай, когда в рамках теста одна из основных метрик значимо ухудшилась, а другая значимо улучшилась. Такой случай может вызвать конфликт в результатах.

Такой случай можно сгенерировать в виде датасета и рассмотреть. Датасет с такой ситуацией генерируется с помощью кода в файле ficha_gen.py и сохраняется в эксельку ficha_ab.xlsx. Импортируем данный датасет и рассмотрим его.

```python
df = pd.read_excel("ficha_ab.xlsx", sheet_name="ficha")
df.describe()
```

Я смоделировала случай, где CTR превышает конверсию (conversion) в тестовой группе. Давайте более детально посмотрим на эти изменения:

```python
mean_session_length_control = df[df['group'] == 'control']['CTR'].mean()
mean_purchases_control = df[df['group'] == 'control']['conversion'].mean()

mean_session_length_test = df[df['group'] == 'test']['CTR'].mean()
mean_purchases_test = df[df['group'] == 'test']['conversion'].mean()

print(f"Контрольная группа - Средняя CTR: {mean_session_length_control}, Среднее конверсии: {mean_purchases_control}")
print(f"Тестовая группа - Средняя CTR: {mean_session_length_test}, Среднее конверсии: {mean_purchases_test}")

improvement_session_length = mean_session_length_test - mean_session_length_control
decline_purchases = mean_purchases_test - mean_purchases_control

print(f"Увеличение CTR в тестовой группе: {improvement_session_length}")
print(f"Снижение конверсии в тестовой группе: {decline_purchases}")
```

Как мы видим по данным конверсия снизилась в тестовой группе на 0,8%, а длина CTR увеличилась на 0.6% 

Такой конфликт в основных метриках может быть связан с тем, что плиточный вид тяжело воспринимается пользователем, что приводит к увеличению времени, проведенного на сайте, но снижению числа покупок.

Увеличение длины сессии может указывать на то, что пользователи тратят больше времени на просмотр контента, но не находят то, что им нужно. Это может быть следствием перегруженности интерфейса или недостаточной ясности в представлении информации.

Также на это могут повлиять приведенные мной выше факторы - Новизна фичи и Привычка старых пользователей.
